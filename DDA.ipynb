```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = max(abs(dx), abs(dy))
    x_inc = dx / steps
    y_inc = dy / steps

    x, y = x1, y1
    points = [(x, y)]

    for _ in range(int(steps)):
        x += x_inc
        y += y_inc
        points.append((x, y))

    return points

# Example usage
x1, y1 = 2, 3
x2, y2 = 10, 8
points = dda(x1, y1, x2, y2)

# Extract x and y coordinates
x_vals, y_vals = zip(*points)

# Plot with connected lines
plt.plot(x_vals, y_vals, marker='o', color='blue', linestyle='-')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("DDA Line Algorithm")
plt.grid()
plt.show()
```
